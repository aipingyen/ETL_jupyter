```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
import queue
import threading
import time
import json
```

```python
brand_list = ['AUDI', 'M-BENZ', 'BMW', 'FORD', 'HONDA', 'LEXUS', 'MAZDA', 'MITSUBISHI',
              'NISSAN', 'PORSCHE', 'SUZUKI', 'SUBARU', 'TOYOTA', 'VOLVO', 'VOLKSWAGEN']
brands=[]
for word in brand_list:
    brands.append(word.lower())

total=0
for brand in brands:
    count=0
    with open('/home/ubuntu/Desktop/yahoo_html/{}.html'.format(brand)) as f:
        data =  f.read()
    soup = BeautifulSoup(data, 'lxml')
    for a in soup.select('.D-ib p a'):
       
        total+=1
        count+=1
        modelDic={}
        modelDic['brand']=brand
        modelDic['model']=a['href'].split(brand)[1].strip('-')
#         pprint(modelDic)
        with open('yahoo_models.txt','a') as f:
            f.write('{}|{}\n'.format(a['href'], json.dumps(modelDic, ensure_ascii=False)))
#     print(count, brand)
# print('total={}'.format(total))
```

```python
# https://tw.autos.yahoo.com/new-cars/model/audi-a4-sedan-2017|{"brand": "audi", "type": "a4-sedan-2017"}
with open('yahoo_models.txt','r') as f:
    data=f.readlines()

    for line in data:
        infos=line.split('\n')[0].split('|')
        url=infos[0]
        typeDic=json.loads(infos[1])
        
        res = requests.get(url)
        res.encoding = 'utf-8'
        soup = BeautifulSoup(res.text, 'lxml') 
        for a in soup.select('.centercol a'):
            typeDic['type']=a.select_one('.title').text
            with open('yahoo_types.txt','a') as f:
                f.write('{}|{}\n'.format(a['href'],json.dumps(typeDic, ensure_ascii=False)))
```
